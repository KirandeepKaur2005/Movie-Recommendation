```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
    text = str(text).lower()
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    words = text.split()
    words = [word for word in words if word not in stop_words]
    words = [lemmatizer.lemmatize(word) for word in words]

    return " ".join(words)
```

```python
df['tags'] = df['tags'].apply(preprocess_text)
```

```python
df['tags'][1]
```

```python
df = df.reset_index(drop=True)
```

```python
# indices = pd.Series(df.index, index=df['title']).drop_duplicates()
indices = pd.Series(df.index, index=df['title'].str.lower()).drop_duplicates()
indices
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=50000, ngram_range=(1,2), stop_words='english')
```

```python
tfidf_matrix = tfidf.fit_transform(df['tags'])
```

```python
tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend(title, n=10):
    title = title.lower()
    matches = df[df['title'].str.lower() == title]
    
    if matches.empty:
        matches = df[df['title'].str.lower().str.contains(title, na=False)]
    if matches.empty:
        return ['Movie not found']

    idx = matches.index[0]

    sim_score = cosine_similarity(tfidf_matrix[idx], tfidf_matrix).flatten()
    similar_idx = sim_score.argsort()[::-1][1:n+1]

    return df['title'].iloc[similar_idx]
```

```python
recommend('robots')
```
